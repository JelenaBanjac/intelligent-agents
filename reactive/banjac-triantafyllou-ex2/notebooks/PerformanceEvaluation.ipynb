# Performance Evaluation of Different Reactive Agent Settings

# Confidence Interval (CI)

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import colorConverter as cc
import pandas as pd
import numpy as np
```

```python
def plot_mean_and_CI(mean, lb, ub, color_mean=None, color_shading=None):
    # plot the shaded range of the confidence intervals
    plt.fill_between(range(mean.shape[0]), ub, lb,
                     color=color_shading, alpha=.2)
    # plot the mean on top
    plt.plot(mean, color_mean)
 
```

```python
def agent_mean_ub_lw(dfs, agent_name, column, limit_time):
    all_values = np.zeros((len(dfs), limit_time))
    for i, df in enumerate(dfs):
        all_values[i, :] = df[df.agent==agent_name].head(limit_time)[column]
    mean = np.median(all_values, axis=0)
    ub = np.max(all_values, axis=0)
    lw = np.min(all_values, axis=0)
    return mean, ub, lw
```

```python
from os import listdir
from os.path import isfile, join

def collect_all_dfs(topology):
    path_to_files = f"../data/{topology}"
    onlyfiles = [f for f in listdir(path_to_files) if isfile(join(path_to_files, f))]
    print(onlyfiles)
    dfs = []
    for file_name in onlyfiles[:3]:
        df = pd.read_csv(f"{path_to_files}/{file_name}", sep=";", 
                         names=["agent", "num_actions", "total_profit", "average_profit", "reward_per_km"])
        dfs.append(df)
    print(f"Number of dfs {len(dfs)}")

    return dfs
```

```python
def get_stats_for(column, topology):
    stats = {}
    dfs = collect_all_dfs(topology)
    for an in agent_names:
        stats[an] = agent_mean_ub_lw(dfs, agent_name=an, column=column, limit_time=10000)
        
    return stats
```

```python
class LegendObject(object):
    def __init__(self, facecolor='red', edgecolor='white', dashed=False):
        self.facecolor = facecolor
        self.edgecolor = edgecolor
        self.dashed = dashed
 
    def legend_artist(self, legend, orig_handle, fontsize, handlebox):
        x0, y0 = handlebox.xdescent, handlebox.ydescent
        width, height = handlebox.width, handlebox.height
        patch = mpatches.Rectangle(
            # create a rectangle that is filled with color
            [x0, y0], width, height, facecolor=self.facecolor,
            # and whose edges are the faded color
            edgecolor=self.edgecolor, lw=3)
        handlebox.add_artist(patch)
 
        # if we're creating the legend for a dashed line,
        # manually add the dash in to our rectangle
        if self.dashed:
            patch1 = mpatches.Rectangle(
                [x0 + 2*width/5, y0], width/5, height, facecolor=self.edgecolor,
                transform=handlebox.get_transform())
            handlebox.add_artist(patch1)
 
        return patch
```

```python
fn = f"../data/switzerland/performance1.csv"
df = pd.read_csv(fn, sep=";", names=["agent", "num_actions", "total_profit", "average_profit", "reward_per_km"])
agent_names = df.agent.unique()
```

```python
%matplotlib inline

stats = get_stats_for("total_profit", topology="switzerland")

colors=['blue', 'green', 'mediumvioletred', 'gold', 'brown', 'red']

fig = plt.figure(1, figsize=(10, 7))
for i, an in enumerate(agent_names):
    plot_mean_and_CI(*stats[an], color_mean=colors[i], color_shading=colors[i])
    
bg = np.array([1, 1, 1])
      
# with alpha = .5, the faded color is the average of the background and color
colors_faded = [(np.array(cc.to_rgb(color)) + bg) / 2.0 for color in colors]
handler_map = {}
for i in range(len(agent_names)):
    handler_map[i] = LegendObject(colors[i], colors_faded[i])
    
plt.legend(list(range(len(agent_names))), agent_names,
           handler_map=handler_map)
plt.xlabel("time")
plt.ylabel("total profit [CHF]")
plt.grid()
plt.show()
```

```python
%matplotlib inline

stats = get_stats_for("reward_per_km", topology="switzerland")

#colors=['mediumvioletred', 'blue', 'brown', 'red', 'gold', 'green']

fig = plt.figure(1, figsize=(10, 7))
for i, an in enumerate(agent_names):
    plot_mean_and_CI(*stats[an], color_mean=colors[i], color_shading=colors[i])
    
bg = np.array([1, 1, 1])
      
# with alpha = .5, the faded color is the average of the background and color
colors_faded = [(np.array(cc.to_rgb(color)) + bg) / 2.0 for color in colors]
handler_map = {}
for i in range(len(agent_names)):
    handler_map[i] = LegendObject(colors[i], colors_faded[i])
    
plt.legend(list(range(len(agent_names))), agent_names,
           handler_map=handler_map)
plt.xlabel("time")
plt.ylabel("reward per km [CHF/km]")
plt.grid()
plt.show()
```
