```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'a': np.arange(12),
        'c': np.random.randint(0, 12, 12),
        'd': np.random.randn(12)}
data['b'] = data['a'] + 10 * np.random.randn(12)
data['d'] = np.abs(data['d']) * 100

plt.figure(figsize=(9, 5))
plt.plot(data['a'], data['b'], marker='o', label='BFS')
plt.plot(data['a'], data['c'], marker='o', label='A*')
plt.title('Optimal travel distance found by BFS and A* algorithms')
plt.xlabel('Number of tasks')
plt.ylabel('Optimal plan distance')
plt.legend()
plt.show()
```
